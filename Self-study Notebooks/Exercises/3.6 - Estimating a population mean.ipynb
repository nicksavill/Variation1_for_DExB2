The males of stalk-eyed flies (*Cyrtadiopsis dalmanni*) from Malaysia have eyes on the ends of long stalks. The females sometimes use the length of these eye stalks to choose mates.

Researchers measured the span, in millimetres, from one eye to the other of 193 male flies. The data are in the file `Datasets/stalkies.csv`.

1. Import the pandas, seaborn and math modules.
2. Read in the csv file `Datasets/stalkies.csv` into a pandas DataFrame called `stalkies`. 
3. Plot a histogram of `stalkies['span']` and add some useful annotation.
4. Calculate and print the sample mean.
5. Calculate and print the standard error of the mean.
6. Calculate and print the lower and upper limits of the 95% confidence interval of the population mean.
7. Create a bar plot of the estimate of the population mean and its precision. 
8. Write a short sentence reporting your estimates of the population mean and its precision.

<div class="alert alert-info">
Your answer should be

    xbar = 8.56 mm
    SEM = 0.03 mm
    lower limit = 8.50 mm
    upper limit = 8.61 mm
</div>

```python
import pandas as pd
import seaborn as sns
import math

stalkies = pd.read_csv('Datasets/stalkies.csv')

g = sns.displot(stalkies['span'])

g.ax.set_xlabel('Eye span (mm)')
g.ax.set_ylabel('Number of flies')
g.ax.set_title('Eye span of 193 male stalk-eyed flies');

n = stalkies['span'].count()   # sample size
xbar = stalkies['span'].mean() # sample mean
s = stalkies['span'].std()     # sample standard deviation

SEM = s / math.sqrt(n) # standard error of the mean

lower_limit = xbar - 2 * SEM   # Lower limit of 95% CI
upper_limit = xbar + 2 * SEM   # Upper limit of 95% CI

print(f"xbar = {xbar:.2f} mm")
print(f"SEM = {SEM:.2f} mm")
print()
print(f'lower limit = {lower_limit:.2f} mm')
print(f'upper limit = {upper_limit:.2f} mm')

# create a bar plot of the estimate and precision of the population mean
g = sns.catplot(data=stalkies, kind='bar')

# Add some annotation
g.ax.set_ylabel('Eye span (mm)')
g.ax.set_title('Estimated eye span (95%CI)');
```

Mean eye span of stalk-eyed flies is 8.56 mm, 95% CI [8.50 - 8.61 mm].

> Report your estimate and precision

Researchers recorded the duration in seconds of the flash of a sample of male fireflies of the species *Photinus ignitus*. The data are in the file `Datasets/firefly flash.csv`.

1. Import the pandas and seaborn modules and sqrt from the math module.
2. Read in the csv file `Datasets/firefly flash.csv` into a pandas DataFrame called `flash`. 
3. Plot a histogram of `flash['duration']` and add some useful annotation.
4. Calculate and print the sample mean.
5. Calculate and print the standard error of the mean.
6. Calculate and print the lower and upper limits of the 95% confidence interval of the population mean.
7. Create a bar plot of the estimate of the population mean and its precision. 
8. Write a short sentence reporting your estimates of the population mean and its precision.

<div class="alert alert-info">
Your answer should be

    xbar = 0.0964 s
    SEM = 0.0012 s
    lower limit = 0.0940 s
    upper limit = 0.0988 s
</div>

```python
%matplotlib inline
import pandas as pd
import seaborn as sns
from math import sqrt    # Import square root function from math

flash = pd.read_csv('Datasets/firefly flash.csv')

g = sns.displot(flash['duration'])

g.ax.set_xlabel('Duration (s)')
g.ax.set_ylabel('Number of fireflies')
g.ax.set_title('Flash duration of 80 male fireflies');


n = flash['duration'].count()   # sample size
xbar = flash['duration'].mean() # sample mean
s = flash['duration'].std()     # sample standard deviation

SEM = s / sqrt(n) # standard error of the mean

lower_limit = xbar - 2 * SEM   # Lower limit of 95% CI
upper_limit = xbar + 2 * SEM   # Upper limit of 95% CI

print(f"xbar = {xbar:.4f} s")
print(f"SEM = {SEM:.4f} s")
print()
print(f'lower limit = {lower_limit:.4f} s')
print(f'upper limit = {upper_limit:.4f} s')

# create a bar plot of the estimate and precision of the population mean
g = sns.catplot(data=flash, kind='bar')

# Add some annotation
g.ax.set_ylabel('Duration (s)')
g.ax.set_title('Estimated flash duration of male fireflies (95%CI)');
```

Mean flash duration of fireflies is 0.0964 s, 95% CI [0.0940 - 0.0988 s].

> Report your estimate and precision
